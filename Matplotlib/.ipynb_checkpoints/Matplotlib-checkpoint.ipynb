```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib notebook
%matplotlib inline
#%matplotlib qt
```

# Matplotlib库的使用

```python
plt.plot([3,1,4,5,2])
plt.ylabel("Grade")
plt.show()
```

**plt.plot()只有一个输入列表或数组时，参数被当作Y轴，X轴以索引自动生成**  

```python
# %matplotlib notebook
plt.plot([5,4,5,1,2])
plt.ylabel("Grade")
plt.savefig("test",dpi = 600)   #PNG文件
plt.show()
```

**plt.savefig()将输出图形存储为文件，默认PNG格式，可以通过dpi修改输出质量**

```python
plt.plot([0,2,4,6,8],[3,1,4,5,2])
plt.ylabel("Grade")
plt.axis([-1, 10, 0, 6])
plt.show()
```

**plt.plot(x,y)当有两个以上参数时，按照X轴和Y轴顺序绘制数据点**  
# Matplotlib库绘图区域

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
a = np.arange(0,5,0.02)
plt.subplot(211)
plt.plot(a,f(a))
plt.subplot(2,1,2)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.show()
```

# Pyplot的文本显示

```python
a = np.arange(0,5,0.02)
plt.xlabel("横轴（时间）",fontproperties = "SimHei",fontsize = 15,color = 'green')
plt.ylabel("纵轴（振幅）",fontproperties = "SimHei",fontsize = 15)
plt.title(r'正弦波实例$y = cos(2\pi x)$',fontproperties = "SimHei",fontsize = 20)
plt.text(2,1,r'$\mu = 100 $',fontsize = 15)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.axis([-1,6,-2,2])
plt.grid(True)
plt.show()
```

**注意Latex公式的使用**

```python
a = np.arange(0,5,0.02)
plt.xlabel("横轴（时间）",fontproperties = "SimHei",fontsize = 15,color = 'green')
plt.ylabel("纵轴（振幅）",fontproperties = "SimHei",fontsize = 15)
plt.title(r'正弦波实例$y = cos(2\pi x)$',fontproperties = "SimHei",fontsize = 20)
plt.annotate(r'$\mu = 100 $',xy = (2,1),xytext = (3,1.5),arrowprops = dict(facecolor = 'blue',shrink = 0.1,width = 2))
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.axis([-1,6,-2,2])
plt.grid(True)
plt.show()
```

# Pyplot子绘图区域

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
a = np.arange(0,5,0.02)
plt.subplot2grid((3,3),(0,0),colspan = 3)
plt.plot(a,f(a))
plt.subplot2grid((3,3),(1,0),colspan = 2)
plt.plot(a,f(a))
plt.subplot2grid((3,3),(1,2),rowspan = 2)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.subplot2grid((3,3),(2,0))
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.subplot2grid((3,3),(2,1))
plt.plot(a,f(a))
plt.show()
```

```python
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3,3)
plt.subplot(gs[0,:])
plt.plot(a,f(a))
plt.subplot(gs[1:,2])
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.show()
```

# pyplot基础图表函数

```python
import matplotlib.pyplot as plt
labels = 'frogs','Hogs','Dogs',"Logs"
sizes = [15,30,45,10]
explode = [0,0.1,0,0]
plt.pie(sizes,explode = explode,labels = labels,autopct = "%.1f%%",shadow = True, startangle = 45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
mu, sigma = 100, 20    #均值和标准差
a = np.random.normal(mu,sigma,size = 100)
plt.hist(a, bins=40, histtype = 'stepfilled',facecolor = 'b', alpha = 0.75)
plt.title("直方图",fontproperties = "SimHei",fontsize = 20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
N = 20
theta = np.linspace(0,2*np.pi,N,endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi/4 * np.random.rand(N)
ax = plt.subplot(111,projection = 'polar')
bars = ax.bar(theta,radii,width = width, bottom=0.0)
for r,bar in zip(radii,bars):
    bar.set_facecolor(plt.cm.viridis(r/10.))
    bar.set_alpha(0.5)
plt.title("雷达图",fontproperties = "SimHei",fontsize = 20)
plt.show()
```
