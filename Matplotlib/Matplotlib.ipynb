# Matplotlib库绘图区域

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
a = np.arange(0,5,0.02)
plt.subplot(211)
plt.plot(a,f(a))
plt.subplot(2,1,2)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.show()
```

# Pyplot子绘图区域

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
a = np.arange(0,5,0.02)
plt.subplot2grid((3,3),(0,0),colspan = 3)
plt.plot(a,f(a))
plt.subplot2grid((3,3),(1,0),colspan = 2)
plt.plot(a,f(a))
plt.subplot2grid((3,3),(1,2),rowspan = 2)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.subplot2grid((3,3),(2,0))
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.subplot2grid((3,3),(2,1))
plt.plot(a,f(a))
plt.show()
```

```python
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3,3)
plt.subplot(gs[0,:])
plt.plot(a,f(a))
plt.subplot(gs[1:,2])
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.show()
```

# 引力波的绘制

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
wavfile.read(r"./grawave/H1_Strain.wav","rb")
```

```python
np.genfromtxt(r'./grawave/wf_template.txt')
```

**读取数据文件**

```python
rate_h, hstrain= wavfile.read(r"./grawave/H1_Strain.wav","rb")
rate_l, lstrain= wavfile.read(r"./grawave/L1_Strain.wav","rb")
reftime, ref_H1 = np.genfromtxt(r'./grawave/wf_template.txt').transpose() #使用python123.io下载文件
```

**将采样率转换为时间间隔**

```python
htime_interval = 1/rate_h
ltime_interval = 1/rate_l
```

**计算数据的时间长度，确定每一个数据点对应的x坐标值**

```python
htime_len = hstrain.shape[0]/rate_h
htime = np.arange(-htime_len/2, htime_len/2 , htime_interval)
htime
```

**绘制第一个引力波**

```python
fig = plt.figure(figsize=(12, 6))
plth = fig.add_subplot(221)
plth.plot(htime, hstrain, 'y')
plth.set_xlabel('Time (seconds)')
plth.set_ylabel('H1 Strain')
plth.set_title('H1 Strain')
plt.show()
```

**绘制第二个引力波**

```python
fig = plt.figure(figsize=(12, 6))
ltime_len = lstrain.shape[0]/rate_l
ltime = np.arange(-ltime_len/2, ltime_len/2 , ltime_interval)
pltl = fig.add_subplot(222)
pltl.plot(ltime, lstrain, 'g')
pltl.set_xlabel('Time (seconds)')
pltl.set_ylabel('L1 Strain')
pltl.set_title('L1 Strain')
plt.show()
```

**绘制理想引力波**

```python
fig = plt.figure(figsize=(12, 6))
pltref = fig.add_subplot(212)
pltref.plot(reftime, ref_H1)
pltref.set_xlabel('Time (seconds)')
pltref.set_ylabel('Template Strain')
pltref.set_title('Template')
fig.tight_layout()
plt.show()
```

**显示多图**

```python
fig = plt.figure(figsize=(12, 6))
htime_len = hstrain.shape[0]/rate_h
htime = np.arange(-htime_len/2, htime_len/2 , htime_interval)
plth = fig.add_subplot(221)
plth.plot(htime, hstrain, 'y')
plth.set_xlabel('Time (seconds)')
plth.set_ylabel('H1 Strain')
plth.set_title('H1 Strain')

ltime_len = lstrain.shape[0]/rate_l
ltime = np.arange(-ltime_len/2, ltime_len/2 , ltime_interval)
pltl = fig.add_subplot(222)
pltl.plot(ltime, lstrain, 'g')
pltl.set_xlabel('Time (seconds)')
pltl.set_ylabel('L1 Strain')
pltl.set_title('L1 Strain')

pltref = fig.add_subplot(212)
pltref.plot(reftime, ref_H1)
pltref.set_xlabel('Time (seconds)')
pltref.set_ylabel('Template Strain')
pltref.set_title('Template')
fig.tight_layout()

plt.savefig("./grawave/Gravitational_Waves.png")
plt.show()
plt.close(fig)
```
